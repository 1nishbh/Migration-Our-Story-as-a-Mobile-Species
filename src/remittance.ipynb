Outward Remittance: Remittance going out from a mentioned country

```python
remitance = pd.read_excel("../data/outward-remittance-flows-brief-39-december-2023-revised-as-of-mar.8-2024_1.xlsx", header = 1, index_col= None)
```

```python
remitance.head()
```

# First Decade (From 2000-2010)

```python
remitance_10 = remitance.iloc[:, 0:12]
remitance_10.head()
```

```python
remitance_10['Mean Remitance'] = remitance_10.iloc[:, 1:].mean(axis=1)
```

```python
remitance_10.head()
```

```python
top_10 = remitance_10.nlargest(10, 'Mean Remitance')
```

```python
top_10
```

# Second Decade (From 2010-2020)

```python
remitance_20 = remitance
remitance_20 = remitance_20.drop(remitance_20.columns[1:12], axis=1)
remitance_20 = remitance_20.iloc[:, :-3]
remitance_20["Mean Remitance"] = remitance_20.iloc[:, 1:].mean(axis=1)
remitance_20.head()
```

```python
top_10_sec = remitance_20.nlargest(10, 'Mean Remitance')
top_10_sec
```

```python
top_10_new = top_10.iloc[:, [0, -1]].copy()
top_10_new.rename(columns={'Mean Remitance': '2000-2010'}, inplace=True)
top_10_new
```

```python
top_10_sec_new = top_10_sec.iloc[:, [0, -1]].copy()
top_10_sec_new.rename(columns = {'Mean Remitance': '2010-2020'}, inplace= True)
top_10_sec_new
```

```python
combined_top_10 = pd.merge(top_10_new, top_10_sec_new, on='Remittance outflows (US$ million)', how='outer')
combined_top_10
```

```python
combined_top_10.fillna(0, inplace=True)
combined_top_10
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(18, 10)) 

combined_top_10.plot(x='Remittance outflows (US$ million)', y=['2000-2010', '2010-2020'], kind='bar', ax=ax, color=['maroon', 'blue'])
ax.set_xlabel('Country Sending Out Remittance')
ax.set_ylabel('Mean Remittance (US$ million)')
ax.set_title('Top 10 Remittance Outflows across Two Decades')
ax.set_xticklabels(combined_top_10['Remittance outflows (US$ million)'], rotation=45, ha="right")
ax.legend(['2000-2010', '2010-2020'])
# plt.tight_layout()
plt.savefig("../results/total_remittance_outflows across two decades.png", bbox_inches='tight')
plt.show()
```

Remittance Inward Flow: Countries getting the highest of remittance from abroad

```python
# remittance inward flow

remit_inward = pd.read_excel("../data/inward_remittance_flows_december_2023_1.xlsx")
```

```python
remit_inward.head()
```

```python
remit_inward.tail()
```

# First Decade (From 2000-2010)

```python
remitance_10_in = remit_inward.iloc[:214, 0:12]
remitance_10_in['Mean Remitance'] = remitance_10_in.iloc[:, 1:].mean(axis=1)
remitance_10_in.head()
```

```python
top_10_in = remitance_10_in.nlargest(10, 'Mean Remitance')
```

```python
top_10_in
```

# Second Decade (From 2010-2020)

```python
remitance_20_in = remit_inward
remitance_20_in = remitance_20_in.drop(remitance_20_in.columns[1:12], axis=1)
remitance_20_in = remitance_20_in.iloc[:, :-4]
remitance_20_in["Mean Remitance"] = remitance_20_in.iloc[:214, 1:].mean(axis=1)
remitance_20_in.head()
```

```python
top_10_sec_in = remitance_20_in.nlargest(10, 'Mean Remitance')
top_10_sec_in
```

```python
top_10_new_in = top_10_in.iloc[:, [0, -1]].copy()
top_10_new_in.rename(columns={'Mean Remitance': '2000-2010'}, inplace=True)
top_10_new_in
```

```python
top_10_sec_new_in = top_10_sec_in.iloc[:, [0, -1]].copy()
top_10_sec_new_in.rename(columns = {'Mean Remitance': '2010-2020'}, inplace= True)
top_10_sec_new_in
```

```python
combined_top_10_in = pd.merge(top_10_new_in, top_10_sec_new_in, on='Remittance inflows (US$ million)', how='outer')
combined_top_10_in.fillna(0, inplace=True)
combined_top_10_in
```

```python
fig, ax = plt.subplots(figsize=(18, 8))

combined_top_10_in.plot(x='Remittance inflows (US$ million)', y=['2000-2010', '2010-2020'], kind='bar', ax=ax, color=['maroon', 'blue'])
ax.set_xlabel('Countries Receiving Remittance')
ax.set_ylabel('Mean Remittance (US$ Million)')
ax.set_title('Top 10 Remittance Inflows across Two decades')
ax.set_xticklabels(combined_top_10_in['Remittance inflows (US$ million)'], rotation=45)
ax.legend(['2000-2010', '2010-2020'])
fig.savefig("../results/top_10_remittance_inflows.png", bbox_inches = 'tight')
plt.show()
```
